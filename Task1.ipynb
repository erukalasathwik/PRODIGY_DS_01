```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = ('/content/sample_data/world_population1.csv')
df = pd.read_csv(file_path)
```

```python
df.head(266)
```

```python
df.describe()
```

```python
# Bar chart for the top 10 most populous countries in 2022
top_10_countries = df.nlargest(10, '2022 Population')

plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country/Territory'], top_10_countries['2022 Population'], color='royalblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar chart for population distribution by continent
continent_population = df.groupby('Continent')['2022 Population'].sum().sort_values()

plt.figure(figsize=(10, 5))
plt.bar(continent_population.index, continent_population.values, color='seagreen')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.title('Total Population by Continent in 2022')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
